## **Wordcloud**
Common Parameters:<br>
**width**, **height**: int, width and height of the word cloud image<br>
**background_color**: color, background color of the word cloud<br>
**colormap**: str or Colormap, colormap for coloring the word cloud<br>
**max_words**: int, maximum number of words to display<br>
**stopwords**: set, set of words to ignore<br>
**mask**: ndarray, mask for the word cloud shape<br>
**contour_width**, contour_color**: int, width and color of the contour line<br>
**random_state**: int or None, seed for random number generation<br>
**scale**: float, scaling factor for the word cloud<br>


```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
text_data = "Word clouds are a popular way to visualize text data and highlight the most frequent words."

# Generate WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis', max_words=100, stopwords=None, contour_width=2, contour_color='black').generate(text_data)

# Display WordCloud using plt.imshow
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off axis labels
plt.title('Word Cloud Example')
plt.show()
```
