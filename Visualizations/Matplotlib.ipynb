```python
import matplotlib.pyplot as plt
```

## Line Plot
#### Common Parameters:<br>
**x**: array-like or scalar, the data for the x-axis<br>
**y**: array-like or scalar, the data for the y-axis<br>
**fmt**: format string, controls the line style, marker, and color<br>
**color**: color of the line<br>
**linestyle** (or ls): line style ('-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted')<br>
**linewidth** (or lw): the width of the line<br>
**marker**: marker style ('.', 'o', 's', '+', '*', 'D', 'x', 'v', '^', '<', '>', '1', '2', '3', '4', etc.)<br>
**markersize** (or ms): size of the marker<br>
**markeredgecolor**: color of the marker edge<br>
**markerfacecolor**: color of the marker face<br>
**label**: label for the legend<br>
**alpha**: transparency of the line<br>

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, color='blue', linestyle='--', marker='o', markersize=8, markerfacecolor='red', markeredgecolor='black', linewidth=2, label='Line 1')
plt.title('Example Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

## Area Plot
Common Parameters:<br>
**x**: array-like or scalar, the data for the x-axis<br>
**y1**: array-like or scalar, the lower curve data for the y-axis<br>
**y2**: array-like or scalar, the upper curve data for the y-axis (default is 0, creating an area plot)<br>
**where**: array-like, boolean mask or condition, where to fill the area<br>
**interpolate**: boolean, whether to interpolate between points<br>
**color**: color of the filled area<br>
**alpha**: transparency of the filled area<br>
**label**: label for the legend<br>

```python
import numpy as np
x = np.linspace(0, 10, 100)
y1 = x**2

plt.fill_between(x, y1, color='red', alpha=0.4, label='Filled Area')
plt.title('Example Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

## Histogram
Common Parameters:<br>
**x**: array-like, the data to be plotted<br>
**bins**: int or array-like, the number of bins or bin edges<br>
**range**: tuple, the lower and upper range of the bins<br>
**density**: boolean, if True, the first element of the return tuple will be the counts <br>normalized to form a probability density<br>
**color**: color or list of colors, the color or colors of the bars<br>
**edgecolor**: color, the color of the histogram edges<br>
**alpha**: scalar, transparency of the bars<br>
**label**: label for the legend<br>

```python
data = np.random.randn(1000)  # Example data

plt.hist(data, bins=30, density=True, color='green', edgecolor='black', alpha=0.7, label='Histogram')
plt.title('Example Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## **2d Histogram**

**range**: array-like, the lower and upper range of the bins along each dimension<br>
**density**: boolean, if True, the histogram is normalized to form a probability density<br>

## **Hexbin**
Common Parameters:<br>
**x**, **y**: array-like, the data for the x and y coordinates<br>
**gridsize**: int or [int, int], the number of hexagons in the x and y directions<br>
**cmap**: Colormap, Colormap object that maps values to colors<br>
**bins**: 'log' or array-like, the bin edges or the type of binning to use ('log' for log-scale<br> binning)<br>
**mincnt**: int, the minimum number of points in a hexagon for it to be colored<br>

```python
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.hexbin(x, y, gridsize=30, cmap='Greens_r', bins='log', mincnt=1)
plt.colorbar(label='Log Count')
plt.title('Example Hexbin Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## Pie Chart<br>
Common Parameters:<br>
**x**: array-like, the data for the pie chart<br>
**labels**: list, labels for each wedge<br>
**colors**: array-like, colors of the wedges<br>
**autopct**: string or function, used to label the wedges with their numeric value<br>
**startangle**: float, the angle by which the start of the pie is rotated<br>
**shadow**: boolean, if True, the pie will be drawn with a shadow<br>
**explode**: array-like, specifies the fraction of the radius with which to offset each wedge<br>
**pctdistance**: float, the ratio between the center of each pie slice and the start of the <br>text generated by autopct<br>


```python
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30,20, 25]

plt.pie(sizes,labels = labels,rotatelabels=True,explode = [0,0,0.04,0],radius=44, autopct='%1.0f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon'], shadow=True)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Example Pie Chart')
plt.show()
```

## Box Plot
Commonly Used Parameters:<br>
**x**: array-like, the input data for the boxplot<br>
**notch**: boolean, if True, will produce a notched box plot<br>
**vert**: boolean, if True, boxes are vertical; if False, boxes are horizontal<br>
**patch_artist**: boolean, if True, will fill the box<br>
**boxprops**: dictionary, properties for the box (e.g., facecolor, color)<br>
**medianprops**: dictionary, properties for the median line (e.g., color)<br>
**flierprops**: dictionary, properties for the fliers (outliers) (e.g., markerfacecolor, marker)<br>
**meanprops**: dictionary, properties for the mean point (e.g., marker, markerfacecolor)<br>
**whiskerprops**: dictionary, properties for the whiskers (e.g., color)<br>
**capprops**: dictionary, properties for the caps on the whiskers (e.g., color)<br>
**positions**: array-like, the positions of the boxes<br>
**widths**: float or array-like, the width(s) of the boxes<br>

```python
data = np.random.randn(100, 3)  # Example data with three groups

plt.boxplot(data, notch=False, vert=True, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='red'), 
            medianprops=dict(color='black'), 
            flierprops=dict(markerfacecolor='green', marker='D'), 
            meanprops=dict(marker='s', markerfacecolor='yellow', markeredgecolor='black'),
            whiskerprops=dict(color='purple'), capprops=dict(color='purple'))

plt.title('Example Box Plot')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()
```

```python
def f(x, y):
    return(1-x/2+x**5+y**3)* np.exp(-x**2-y**2)
#Filled
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='jet')
plt.colorbar()
plt.title('Example Filled Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## **3d_Surface**
Common Parameters for plot_surface:<br>
**X**, **Y**, **Z**: 2D arrays representing the x, y, and z coordinates of the surface<br>
**rstride**, **cstride**: int, stride used to sample the input data to plot the surface<br>
**cmap**: Colormap, Colormap object that maps values to colors<br>
**facecolors**: array-like, the face colors of the surface<br>
**edgecolor**: color or None, the color of the edges of the surface patches<br>
**linewidth**: float, the width of the surface patch edges<br>
**antialiased**: boolean, if True, the surface will be antialiased<br>


```python
# Example:
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()

# Subplot with projection='3d'
ax = fig.add_subplot(111, projection='3d')

# Surface Plot with Most Used Parameters
surf = ax.plot_surface(X, Y, Z, rstride=5, cstride=5, cmap='viridis', edgecolor='k', linewidth=0.5, antialiased=True)

ax.set_title('Example 3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.view_init(-100,80)
fig.colorbar(surf, ax=ax, label='Z Value')

plt.show()
```

## **3d Line**
Common Parameters for plot3D:<br>
**x**, **y**, **z**: array-like, the data for the x, y, and z coordinates<br>
**label**: str, label for the line in the legend<br>
**color**: color, the color of the line<br>
**linestyle**: str, the style of the line ('-', '--', ':', etc.)<br>
**linewidth**: float, the width of the line<br>
**markersize**: float, the size of markers if markers are used<br>
**marker**: str, the marker style ('o', '^', 's', etc.)<br>
**alpha**: float, the transparency of the line<br>


```python
# Example:
t = np.linspace(0, 10, 100)
x = np.cos(t)
y = np.sin(t)
z = t

# Create a 3D line plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(x, y, z, label='3D Line Plot', color='blue', linestyle='-', linewidth=2, markersize=1, marker='o', alpha=0.8)

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Line Plot Example')

# Add a legend
ax.legend()

plt.show()
```

## **Trisurf**
Common Parameters for plot_trisurf:<br>
**x**, **y**, **z**: array-like, the data for the x, y, and z coordinates<br>
**cmap**: Colormap, Colormap object that maps values to colors<br>
**linewidth**: float, the width of the lines connecting the triangles<br>
**edgecolor**: color or 'none', the color of the edges of the triangles<br>
**antialiased**: boolean, if True, the surface will be antialiased<br>


```python
import matplotlib.pyplot as plt
import numpy as np



# Example:
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = r * np.sin(theta)
y = r * np.cos(theta)
z = f(x, y)

# Plot the trisurf without explicitly specifying triangles
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_trisurf(x, y, z, cmap='twilight', linewidth=0.2, edgecolor='black', antialiased=True)

ax.set_title('3D Triangular Surface Plot (Automatic Triangulation)')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.show()
```

## **3d Contour**
Common Parameters for contour3D:<br>
**x**, **y**, **z**: array-like, the data for the x, y, and z coordinates<br>
**cmap**: Colormap, Colormap object that maps values to colors<br>
**levels**: int or array-like, the number of contour levels or the specific levels<br>
**linewidths**: float or array-like, the width of the contour lines<br>
**linestyles**: str or array-like, the style of the contour lines<br>
**alpha**: float, the transparency of the contours<br>


```python
# Example:
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D contour plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
contour = ax.contour3D(X, Y, Z, cmap='ocean_r', levels=20, linewidths=2, linestyles='solid', alpha=0.8)

ax.set_title('3D Contour Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Add color bar
fig.colorbar(contour, ax=ax, label='Z Value')

plt.show()
```
